```python
s = ''
c = cible['Station'][1]
for i, r in cible.iterrows():
    if r['Station'] != c:
        s += '2=\n'
        c = r['Station']
    s += '2=' + r['Station'] + '\n' + '5=' + r['Cible'] + '\n' + '7=' + str(r['lecture']) + '\n' + '8=100' + '\n' + '9=' + str(r['distance']) + '\n' 
```

```python
text_file = open("sample.txt", "w")
n = text_file.write(s)
text_file.close()
```

```python
f = open("test-1.prm", "w")
n = f.write(s)
f.close()

lines_seen = set() # holds lines already seen
with open("test-1.prm", "r+") as f:
    d = f.readlines()
    f.seek(0)
    for i in d:
        if i.startswith('2='):
            if i == '2=\n': f.write(i)
            elif (i not in lines_seen):
                f.write(i)
                lines_seen.add(i)
        else:
            f.write(i)
                
    f.truncate()
```
